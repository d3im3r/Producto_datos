```python
# Manipulacion de datos...
import pandas as pd
import numpy as np
# Visualizacion de datos...
import matplotlib.pyplot as plt
import seaborn as sns
# No presentar advertencias......
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../Data/Original/hotel_bookings.csv")
df.head()
```

```python
df["arrival_date_month"].unique()
```

```python
def monthToNum(shortMonth):
    return {
            'January': 1,
            'February': 2,
            'March': 3,
            'April': 4,
            'May': 5,
            'June': 6,
            'July': 7,
            'August': 8,
            'September': 9, 
            'October': 10,
            'November': 11,
            'December': 12
    }[shortMonth]
```

```python
df["arrival_date_month"] = df["arrival_date_month"].apply(monthToNum)
df.head()
```

```python
df['Arrival_date']=pd.to_datetime({'year': df["arrival_date_year"],
                           'month': df["arrival_date_month"],
                           'day': df["arrival_date_day_of_month"]
                           })

df.head()
```
