```python
#importar librerías
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
#cargar datos
datos=pd.read_csv("adult.data",header=0)
datos.head()
```

```python
#selección de 10k datos aleatorios
data = datos.sample(n=10000, random_state=13)
data.head()
```

```python
#Busco datos nulos o vacíos
data.isnull().sum()
```

```python
#Análisis exploratorio
data.describe().T
```

```python
#Gráficos estadísticos de los datos (correlación por pares)
plt.figure(figsize=(10,5))
sns.pairplot(data=data, hue="income")
plt.title("Gráfico correlaciones por pares")
plt.show()
```
